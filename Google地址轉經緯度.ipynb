```python
def visit():
    driver.get(url);
    
def Search_grep():
    try:       
        filename = r'C:\Users\student\Downloads\地址輸出.xlsx'

        data = pd.read_excel(filename)

        result = []

        # 将每个单元格的数据转换为列表
        for row in data.values:
            row_list = []
            for cell in row:
                row_list.append(cell)
                result.append(row_list)

        # 在循环中处理列表数据，并写入"地點"
        for row in result:
            for cell in row:
        
                
                
                sleep(1)

                spot = driver.find_element(By.ID, "searchboxinput")
                spot.send_keys(cell)

                search = driver.find_element(By.ID, "searchbox-searchbutton")
                search.click()

                sleep(5)

                current_url = driver.current_url
                latitude, longitude = current_url.split("@")[1].split(",")[:2]
                print(latitude, longitude)

                filename = r"C:\Users\student\python_web_scraping\經緯度.xlsx"

                workbook = openpyxl.load_workbook(filename)
                worksheet = workbook.worksheets[0]

                next_row = worksheet.max_row + 1

                cell1 = worksheet.cell(row=next_row, column=1)
                cell1.value = cell

                cell2 = worksheet.cell(row=next_row, column=2)
                cell2.value = float(longitude)

                cell3 = worksheet.cell(row=next_row, column=3)
                cell3.value = float(latitude)

                workbook.save(filename)

                print("完成")
                
                sleep(2)
            
                spot.clear()
        
        return     
        
    except TimeoutException:
        print("等待逾時，即將關閉瀏覽器…") 
   
```

```python
if __name__ == '__main__':
    visit()
    Search_grep()
```
